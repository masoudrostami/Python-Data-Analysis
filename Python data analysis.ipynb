## Data Wrangling

```python
## choose the column you want

new_covid_data = covid_data[['date','state','death','positive','negative','totalTestResults']]
new_covid_data.head(10) 
```

```python
new_covid_data.shape
```

### Renmae the column variable names

```python
new_covid_data.rename({'death':'dead'},axis=1,inplace=True) # axis = (0 = rows, 1 = columns), inplace=True replaces the current DataFrame
```

```python
new_covid_data.head()
```

### Reorder the columns

```python
## 1) First methos

new_covid_data = new_covid_data[['date','state','totalTestResults','negative','positive','dead']]

```

```python
## 1) Second methos

new_covid_data = new_covid_data.iloc[:,[0,1,5,4,3,2]] #iloc is how you access df like a matrix
```

```python
### Adding a new variable

new_covid_data['rate'] = new_covid_data.positive / new_covid_data.totalTestResults #alternate way
```

```python
new_covid_data.head() # variable rate added
```

```python
#### working with dates
```

```python
new_covid_data["date"]= pd.to_datetime(new_covid_data["date"],yearfirst=True) 
new_covid_data.info()
```

```python
new_covid_data.describe()
```

```python
### add a column for day of year
```

```python
new_covid_data['dayofyear'] = new_covid_data['date'].dt.dayofyear
new_covid_data.head()
```

```python
## week of the year
new_covid_data['dayofyear'] = new_covid_data['date'].dt.week
new_covid_data.head()
```

### Filtering the dataset

```python
new_covid_data.state== "AZ"
```

```python
AZ_covid_data = new_covid_data[new_covid_data.state == 'AZ']
AZ_covid_data.head()
```

```python
###
```

```python
new_covid_data[(new_covid_data.state == 'AZ') & (new_covid_data.dead >15600)]
```

### using df.query()

df.query() is another way that might look a little cleaner for more complicated 

```python
new_covid_data.query("state == 'AZ' & rate > 0.12")
```

### Sorting

```python
new_covid_data.sort_values('positive',ascending=False,inplace=True)
new_covid_data.head()
```

```python
### Grouping state variable (median)

covid_state = new_covid_data.groupby('state',as_index=True).agg(['median','count'])
covid_state
```

```python
### Grouping state variable (mean)

covid_state2 = new_covid_data.groupby('state',as_index=True).agg(['mean','count'])
covid_state2
```

# 3) matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Make the graphs a bit prettier, and bigger
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
#### plot deaths over time ####
AZ_covid_data.dead.plot()
```

```python
#### deaths by day of the week ####

weekday_counts = AZ_covid_data.groupby(AZ_covid_data['date'].dt.day_name()).aggregate(sum)
weekday_counts
```

```python
weekday_counts.dead.plot(kind='bar')
```

```python




weekday_counts[weekday_counts.dead == weekday_counts.dead.max()]
```
